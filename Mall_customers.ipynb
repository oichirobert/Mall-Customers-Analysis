## **Clustering: Univariate, Bivariate & Multivariate**

# **KMeans Algorithm**

```python
df.columns
```

```python
cluster1 = KMeans(n_clusters=3)
```

```python
cluster1.fit(df[["Annual Income (k$)"]])
```

```python
cluster1.labels_
```

```python
df["Income Cluster"] = cluster1.labels_
```

```python
df.head()
```

```python
df["Income Cluster"].value_counts()
```

```python
cluster1.inertia_
#It shows the distance from each cluster
```

**To Optimise clusters**

```python
inertia_scores =[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(df[["Annual Income (k$)"]])
  inertia_scores.append(kmeans.inertia_)
  
plt.plot(range(1,11), inertia_scores)  
plt.xlabel("Number of clusters")
```

From the above graph we pick 3 clusters, that is where it starts looking like an elbow

**Analysis**

```python
df.columns
```

```python
df.groupby("Income Cluster")['Age', 'Annual Income (k$)',
       'Spending Score (1-100)'].mean()
```

# **Bivariate Analysis**

```python
cluster2 = KMeans(n_clusters=5)
cluster2.fit(df[['Annual Income (k$)',
       'Spending Score (1-100)']])
df['Income_Spending Cluster']=cluster2.labels_
```

```python
df.head()
```

In cluster three, which is the target cluster from the analysis is led by female.

**Bivariate Analysis**

```python
df.groupby("Income_Spending Cluster")['Age','Annual Income (k$)',
       'Spending Score (1-100)'].mean()
```

**Finding the best number of clusters for the data**

```python
inertia_scores2 = []
for i in range(1,10):
  kmeans2=KMeans(n_clusters=i)
  kmeans2.fit(df[['Annual Income (k$)',
       'Spending Score (1-100)']])
  inertia_scores2.append(kmeans2.inertia_)

plt.plot(range(1,10), inertia_scores2);
```

From the above graph, it is evident, 5 clusters selected, since it is where the elbow like curve appears. 

# **Scatter Graph**

Detrmining centers of clusters

```python
centers =pd.DataFrame(cluster2.cluster_centers_)
centers.columns =["x","y"]
```

# **Plotting scatter chart with centers**

```python
plt.figure(figsize=(10,8))
plt.scatter(x=centers["x"], y=centers["y"], s=100, c="black", marker="*")
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue="Income_Spending Cluster", palette="tab10");

#Saving the graph
plt.savefig("Clustering_Bivariate.png")
```
